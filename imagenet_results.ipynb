## Comparison of Low rank approximation of gradients to momentum

```python
# Imports
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# load data
low_rank_no_decay_file = "data/trial_1/eval_measurements.csv"
low_rank_with_decay_file = "data/trial_2/eval_measurements.csv"
control_file = "data/control_trial/eval_measurements.csv"
trial_files = [low_rank_with_decay_file, low_rank_no_decay_file, control_file]
trials = [pd.read_csv(file) for file in trial_files]
names = ["Imagenet resnet low rank with weight decay",
         "Imagenet resnet low rank with no weight decay",
         "Imagenet resnet (momentum)"]
```

```python
# Relevant columns and colors for display
eval_cols =  ["test/loss", "validation/loss", "global_step", "score"]
colors = ["b-", "r-", "m-"]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=1)
for (color, trial, name) in zip(colors, trials, names):
    ax.plot("score", "validation/accuracy", color, data=trial, label=name)
ax.set(xlabel="Time", ylabel="validation/accuracy", title="Low rank tuning trials")
ax.legend()
ax.set_box_aspect(7.0 / 16)
fig.savefig("score_vs_val_acc.png", transparent=True)
```
